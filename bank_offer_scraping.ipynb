```
import pandas as pd 
import numpy as np
from bs4 import BeautifulSoup 
import requests
```

```
url="https://happycredit.in/kotak-mahindra-bank-offers/"
```

```
page=requests.get(url)
```

```
page.text
```

```
soup=BeautifulSoup(page.text,"html.parser")
```

```
soup.title
```

```
name=[]
for a in soup.findAll('a',href=True,attrs={'class':'coupon-cart others'}):
  ni=a.find('div',attrs={'class':'right-clm'})
  name.append(ni.text)
  

```

```
bank_name=[]
for img in soup.findAll('img'):
  bank=img.get('alt')
  bank_name.append(bank)
```

```
df=pd.DataFrame(list(zip(name,bank_name)))
df.columns=['Offer','Bank_name']
```
